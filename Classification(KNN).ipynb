# K-nearest neighbor

1. Import
2. Define นิยายว่าโมเดลจะหน้าตาอย่างไร
3. Train (fit) สร้างโมเดล
4. Test ทำนาย
5. Evaluate เป็นการประเมิน เป็นขั้นตอนของงานวิจัย

```
#Import
from sklearn.neighbors import KNeighborsClassifier
```

```
#Define
myKNN = KNeighborsClassifier(n_neighbors=3)
```

```
#Train
myKNN.fit(X_train,y_train)
```

```
myKNN.predict_proba(X_test)
```

#Selecting the best model

```
from sklearn.model_selection import train_test_split
```

1. แบ่งข้อมูลเป็น train กับ test แล้วเก็บ test ไว้

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

2. ทำ k-fold cross validation เพื่อเลือก model ที่ดีที่สุด

```
from sklearn.model_selection import cross_val_score
```

```
#1. Define model
KNN_k3_wU = KNeighborsClassifier(n_neighbors=3)
#2. Cross validation
cross_val_score(KNN_k3_wU,X_train,y_train,cv=5).mean()
```

```
#1. Define model
KNN_k3_wU = KNeighborsClassifier(n_neighbors=6)
#2. Cross validation
cross_val_score(KNN_k3_wU,X_train,y_train,cv=5).mean()
```

```
#1. Define model
KNN_k3_wU = KNeighborsClassifier(n_neighbors=20,weights='distance')
#2. Cross validation
cross_val_score(KNN_k3_wU,X_train,y_train,cv=5).mean()
```

3. train model ที่ดีที่สุดด้วย training set แล้ววัดผลด้วย test set

```
#Define
KNN_k3_wU = KNeighborsClassifier(n_neighbors=20,weights='distance')
#Train
KNN_k3_wU.fit(X_train,y_train)
#Train_Evaluation
KNN_k3_wU.score(X_test,y_test)
```
